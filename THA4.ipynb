## Measuring violence in the Syria conflict:

To measure levels of violence in the Syria conflict, the following data will be used. Each data set compiles and analyzes violent events with a slightly different definition. 

\
##Data Sources:

### Center for Systemic Peace (CSP) - Major Episodes of Political Violence, 1946 - 2016 ([INSCR](http://www.systemicpeace.org/inscrdata.html))

In the data presented by CSP, "major episodes of violence (MEPV) are defined  by the systematic and sustained use of lethal violence by organized groups that result in at least 500 directly-related deaths over the course of the episode". 
* Episodes of violence are categorized based on the type of conflict: international violence, international war, international independence war, civil violence, civil war, ethnic violence, and ethnic war. 

> * INSCR identifies the conflict in Syria as an "ethnic war"

* Events are framed temporally, and by the extent to which they impact civil society; from a ratio of the two, the "magnitude of impact" score is derived. 

* Events are given magnitude scores ranging from 1 (lowest; "Sporadic or Expressive Political Violence") to 10 (highest; "Extermination and Annihilation"). A score of 0 indicates no event took place. If more than one type of event of violence occured, the magnitude scores are summed. 

> * Syria is scored with a magnitude of 6 for ethnic war, meaning "extensive warfare"

* Total summed scores are provided for societal MEPV, and a combined societal and interstate MEPV are provided and will be included in this research. 

###Armed Conflict Location & Evnt Data ([ACLED](https://www.acleddata.com/data/)) 
ACLED categorizes event types in six different ways. This assignment will focus on only violent events which include: \
1) Battles: "Violent interactions between two organized groups"

2) Explosions/Remote violence: "One-sided violent events in which the tool for engaging in conflict creates asymmetry by taking away the ability of the target to respond"

3) Violence against civilians:* "violent events where an organized armed group deliverately inflicts violence upon unarmed non-combatants"

*Citation:* 

ACLED. (2017). “Armed Conflict Location & Event Data Project (ACLED) Codebook, 2017.”


### Strauss Center - Global Fragility Data Set

**Data Source: Strauss Center State Fragility Initiative [website](https://www.strausscenter.org/strauss-articles/state-fragility.html). The data available for download under the "Data and Maps" section. **

The Strauss Center at LBJ compiles an excellent data set that calculates rates of instability in a given state in a given year. The project employs a comprehensive lens to observing instability and its sources, focusing on states put more at risk because of climate hazards (relevant, because some claim that enviornmental factors played a role in triggering the Syrian conflict). Along with analyzing climate risk factors, the data uses political, security, econmic, and social indicators to measure state fragility. 
* scores presented are "assessed relative to other countries"
* a total fragility score is compiled based on the sum of all 24 idicators used
> * countries are then ranked based on their total fragility score into five categories: low fragility(0 - countries falling below the median score), some (1), moderate (2), high (3), and highest fragility (4) (these four are above the median).



*Citation:* 

Roudabeh Kishi and Andrew Linke. 2016. Global Fragility Dataset, Produced for the U.S. Agency for International Development, Office of Conflict Management and Mitigation. Austin: Robert Strauss Center for International Security and Law.
